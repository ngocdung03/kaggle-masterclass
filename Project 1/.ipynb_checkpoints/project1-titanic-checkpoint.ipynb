### Preprocessing

```python
train.columns.values
```

```python
train.head()
# drop string cols and PassengerId
```

```python
# Handle missing values
train.info()
```

```python
# This is used to determine the list items to include in the titles list in the next cell
train['Name'].unique().tolist()
```

```python
# Extract the title of the person and returns it
titles = ['Mrs', 'Mr', 'Don', 'Jonkheer', 'Master', 'Miss', 'Major',
         'Rev', 'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess']

def get_title(name):
    for title in titles:
        if title in name:
            return title
    return 'None'
print(get_title('Johnson, Mrs. Oscar W (Elisabeth Vilhelmina Berg)'))
print(get_title('Williams, Mr. Charles Eugene'))
```

```python
train['Title'] = train['Name'].apply(lambda x: get_title(x))
train.head()
```

```python
train['Age'].mean()
```

```python
train['Age'].fillna(train['Age'].mean(), inplace = True)
train.head()
```

```python
train.drop('Cabin', axis=1, inplace = True)
train.head()
```

```python
# Embarked has 2 missing values and is a categorical column, therefore:
train.dropna(inplace = True)
train.info()
```

```python
train.head()
# Select columns that are useful to train our model
x_train = train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Title']]
y_train = train['Survived']
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_train = pd.get_dummies(x_train)  #one hot encoding
```

```python
x_train.head()
```

```python
x_train.info()
```

### Model Training

```python
# Initializing the Model
clf = RandomForestClassifier(max_depth=75, random_state=0)
```

```python
# Training the Model
clf.fit(x_train, y_train)
```

```python
# Preprocessing the Test set
test['Title'] = test['Name'].apply(lambda x: get_title(x))
test['Age'].fillna(test['Age'].mean(), inplace = True)
test.drop('Cabin', axis = 1, inplace = True)
test.head()
```

```python
test.info()
```

```python
# Handle one missing value in Fare
test['Fare'].fillna(test['Fare'].mean(), inplace = True)
test.info()
```

```python
x_test = test[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Title']]
x_test.head()
```

```python
x_test = pd.get_dummies(x_test)
```

```python
x_test.info()
```

```python
# Since test set have less unique titles than train set, we need to set the remain values to zero
# Otherwise, errors happen.
total_titles = train['Title'].unique().tolist()
test_titles = test['Title'].unique().tolist()

for title in total_titles:
    if title not in test_titles:
        print('Title ' + str(title) + ' not in Test Set')
        x_test['Title_' + str(title)] = 0
```
